```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Указываем URL страницы
url = "https://fortrader.org/quotes;"

# Загружаем страницу
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Находим все таблицы на странице
tables = soup.find_all('table')

# Импортируем каждую таблицу в DataFrame
for i, table in enumerate(tables, 1):
    # Преобразуем HTML-таблицу в DataFrame
    df = pd.read_html(str(table))[0]

    # Показываем первые 5 строк таблицы
    display(df.head(5))

    # Сохраняем таблицу в отдельный файл CSV
    df.to_csv(f'table_{i}.csv', index=False)
    print(f"Таблица сохранена как 'table_{i}.csv'")
    print("="*50)
```
