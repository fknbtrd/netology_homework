```python
import pandas as pd
from google.colab import files

uploaded = files.upload()
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

fives = ratings[ratings['rating'] == 5.0]

# Считаем, сколько раз каждый фильм получил 5.0
count_fives = fives['movieId'].value_counts().reset_index()
count_fives.columns = ['movieId', 'count_5_stars']

# Добавляем названия фильмов
result = pd.merge(count_fives, movies, on='movieId')

#Сортируем от большего к меньшему
result_sorted = result.sort_values('count_5_stars', ascending=False)

# Берём самый первый фильм из отсортированного списка
most_5_stars_movie = result_sorted.iloc[0]

print("Фильм с наибольшим количеством оценок 5.0:")
print("Название:", most_5_stars_movie['title'])
print("Количество пятёрок:", most_5_stars_movie['count_5_stars'])
```

```python
from google.colab import files
import pandas as pd

# Загружаем файл с компьютера
uploaded = files.upload()
power = pd.read_csv('power.csv')

# Выбираем страны Прибалтики и категории
baltic_countries = ['Latvia', 'Lithuania', 'Estonia']
categories = [4, 12, 21]

# Фильтруем данные
filtered_data = power[
    (power['country'].isin(baltic_countries)) &  # Страны Прибалтики
    (power['category'].isin(categories)) &       # Нужные категории
    (power['year'] >= 2005) & (power['year'] <= 2010) &  # Период 2005-2010
    (power['quantity'] > 0)                     # Только положительные значения
]

# Считаем суммарное потребление
total_consumption = filtered_data['quantity'].sum()

print(f"Суммарное потребление стран Прибалтики: {total_consumption}")
```
